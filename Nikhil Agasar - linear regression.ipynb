```python
# Now we will fill the null values in the dataframe

#Mean is to get the average in an even distribution

df['CRIM'].fillna(df['CRIM'].mean(), inplace=True)

df['ZN'].fillna(df['ZN'].mean(), inplace=True)

df['INDUS'].fillna(df['INDUS'].mean(), inplace=True)

# Mode is to get the most repeated element in the set of i

df['CHAS'].fillna(df['CHAS'].mode()[0], inplace=True)
# Median is often preferred for skewed distributions

df['AGE'].fillna(df['AGE'].median(), inplace=True) 
df['LSTAT'].fillna (df ['LSTAT'].median(), inplace=True)
```

```python
# we will the data tyoe of 'CHAS' column from 
df['CHAS'] = df['CHAS'].astype('int') # exlicit type conver
```

```python
# Univariate distribution of each features
for i in df.columns:
    plt.figure(figsize=(6,3))

    plt.subplot(1, 2, 1)
    df[i].hist(bins=20, alpha=0.5, color='b', edgecolor='black')
    plt.title(f'Histogram of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    plt.boxplot(df[i], vert=False)
    plt.title(f'Boxplot of {i}')
    plt.show()
```

```python
#Correlation Heatmap for each feature against another
corr = df.corr(method='pearson')

plt.figure(figsize=(10, 8))

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.xticks(rotation=90, ha='right')

plt.yticks (rotation=0)

plt.title("Correlation Matrix Heatmap")

plt.show()
```

```python
#  Here X implies  the independent features
X = df.drop('MEDV', axis=1) # All columns expect 'MEDV'
# Here Y implies the dependent features on X values
Y = df['MEDV'] # Target variable
```

```python
#Scole the features 
scale = StandardScaler() 
#We will pass X values inside Standard Scaler, 
#to scale the values
X_scaled = scale.fit_transform(X)
```

```python
#Split the data into training (88%) and testing (20%) set 
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, Y, 
test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
# Initialize the linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)
```

```python
#predict y values on the test set's x values
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate R-squared value
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
